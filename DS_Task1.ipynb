```python
df[df['Country Name']=="India"]
```

```python
cd = df[df['Country Name'] == 'India']

years = cd.columns[4:].astype(int)
population = cd.iloc[0, 4:].astype(int)

plt.figure(figsize=(16, 8))
plt.xlabel('Years',fontsize=12)
plt.ylabel('Population',fontsize=12)
plt.title(f'Population of India Over Years (1960 - 2020)',fontsize=16)

plt.bar(years,population,edgecolor='black')
plt.show()
```

```python
population_in_2023 = df[['Country Name', '2023']]
top_10_countries_in_2023 = population_in_2023.sort_values(by='2023', ascending=False).head(10)
low_10_countries_in_2023 = population_in_2023.sort_values(by='2023' , ascending=True).head(10)
```

```python
top_10_countries_in_2023 
```

```python
low_10_countries_in_2023
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_countries_in_2023 ['Country Name'], top_10_countries_in_2023['2023'], color='red')
plt.xlabel('Population in 2023')
plt.title('Top 10 Countries by Population in 2023')
plt.gca().invert_yaxis()  
plt.show()


plt.figure(figsize=(10, 6))
plt.barh(low_10_countries_in_2023['Country Name'], low_10_countries_in_2023['2023'], color='green')
plt.xlabel('Population in 2023')
plt.title('Lowest 10 Countries by Population in 2023')
plt.gca().invert_yaxis() 
plt.show()
```
